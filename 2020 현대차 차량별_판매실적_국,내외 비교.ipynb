```python
df_modyy = df_mody.dropna()
df_modyy
```

```python
Model_list = df_modyy['Model'].unique().tolist()
```

```python
for i in Model_list:
    df_car = df_modyy[df_modyy['Model']== i]
    nums = df_car.index.to_list()
    df_car=df_car.transpose()
    
    if len(nums) >= 2:
        df_car=df_car.rename(columns={nums[0]:'Domestic',  nums[1]:'Export'})
    else:
        if nums[0] < 50:
            df_car=df_car.rename(columns={nums[0]:'Domestic'})
        else :
            df_car=df_car.rename(columns={nums[0]:'Export'})
    if i != 'Sub-total':
        df_car[1:-1].plot()
        plt.title(i)
        plt.figure(figsize=(15,10))
```
